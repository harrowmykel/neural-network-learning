```python
# get the data without the result
structure_data_shuffled = structure_data_shuffled_df.to_numpy()
structure_data_x_data = (structure_data_shuffled_df.drop(['label'], axis=1)).to_numpy()
structure_data_y_data = (structure_data_shuffled_df['label']).to_numpy()
```

```python
#split the data into test, trin and validation data
X_train, y_train = structure_data_x_data[:30000], structure_data_y_data[:30000] 
X_val, y_val = structure_data_x_data[30000:], structure_data_y_data[30000:] 
columns_count = len(X_train[0])

X_train.shape, y_train.shape, X_val.shape, y_val.shape , type(X_train)
```

```python
# preprocessing with standard deviation to normalise data
from sklearn.preprocessing import StandardScaler
import numpy as np
import matplotlib.pyplot as pyplot
scaler = StandardScaler().fit(X_train)

def preprocessor(X):
  A = np.copy(X)
  A = scaler.transform(A)
  return A

# X_train = preprocessor(X_train)
X_train, X_val = preprocessor(X_train), preprocessor(X_val)

X_train[:20, 456:466]
```

```python
from sklearn.metrics import mean_squared_error as mse
from sklearn.metrics import root_mean_squared_error as rmse

#create a new array that contains every columns in xdata, ydata, and pred
def merge_data(X_data, Y_data, Y_data_prediction):
    index =0
    result = []
    l_rows_count, l_columns_count = X_data.shape
    for element in Y_data:
        new_data =[]
        for i in range(l_columns_count):
            new_data.append( X_data[index, i])
        new_data.append('ans=')
        new_data.append(Y_data[index])
        new_data.append('pred=')
        new_data.append(Y_data_prediction[index])
        result.append(new_data)
        index+=1
    return result

def get_accuracy(predictions, Y):
    return np.sum(predictions == Y) / Y.size

def get_predictions(A2):
    return np.argmax(A2, 0)

def visualise_accuracy(X_train, X_val, y_train_onehot, y_val_onehot, y_train_pred, y_val_pred):
    #visualisation
    X_visual=X_val
    Y_visual=y_val
    Y_visual_pred =y_val_pred
    
    print(get_accuracy(y_val_pred, y_val_onehot))
    print(get_accuracy(y_train_onehot, y_train_pred))
    print(y_val_pred[40], y_val_onehot[40])
    print(y_val_pred[68], y_val_onehot[68])
    print(y_train_pred[40], y_train_onehot[40])
    
    # pyplot.scatter(X_visual[:20, 456], Y_visual[:20], c="red")
    # pyplot.scatter(X_visual[:20, 456], Y_visual_pred[:20], c="blue")
    # lsit = merge_data(X_visual[:20, 463:466],Y_visual[:20], Y_visual_pred[:20])
    # print(pd.DataFrame(lsit))
    # print("\nRMSE:", 'train:', rmse(y_train_pred, y_train), ', valid:', rmse(y_val_pred, y_val))


def visualise(X_train, X_val, y_val, y_train_pred, y_val_pred):
    #visualisation
    X_visual=X_val
    Y_visual=y_val
    Y_visual_pred =y_val_pred
    
    pred = get_predictions(y_val_pred)
    print(get_accuracy(y_val_pred, Y_visual))
    print(get_accuracy(y_val_pred, Y_visual))
    print(y_val_pred[40], Y_visual[40])
    print(y_val_pred[68], Y_visual[68])
    print(y_train_pred[40], Y_visual[40])
    
    pyplot.scatter(X_visual[:20, 456], Y_visual[:20], c="red")
    pyplot.scatter(X_visual[:20, 456], Y_visual_pred[:20], c="blue")
    lsit = merge_data(X_visual[:20, 463:466],Y_visual[:20], Y_visual_pred[:20])
    print(pd.DataFrame(lsit))
    print("\nRMSE:", 'train:', rmse(y_train_pred, y_train), ', valid:', rmse(y_val_pred, y_val))

def onehot(Y):
    one_hot_Y = np.zeros((Y.size, Y.max() + 1))
    one_hot_Y[np.arange(Y.size), Y] = 1
    # one_hot_Y = one_hot_Y.T
    return one_hot_Y
```

```python
y_train_onehot = onehot(y_train)
y_val_onehot = onehot(y_val)

pd.DataFrame(X_train).head()
```

```python
# try to train with different training models
from sklearn.linear_model import RidgeClassifier

lm = RidgeClassifier().fit(X_train, y_train)
#prediction
y_train_pred=lm.predict(X_train)
y_val_pred=lm.predict(X_val)

#visualisation
visualise(X_train, X_val, y_val, y_train_pred, y_val_pred)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=10).fit(X_train, y_train)
#prediction
y_train_pred=knn.predict(X_train)
y_val_pred=knn.predict(X_val)

#visualisation
visualise(X_train, X_val, y_val, y_train_pred, y_val_pred)
```

```python
from sklearn.ensemble import RandomForestClassifier

rfr = RandomForestClassifier(max_depth=10, n_estimators=1000).fit(X_train, y_train)
#prediction
y_train_pred=rfr.predict(X_train)
y_val_pred=rfr.predict(X_val)

#visualisation
visualise(X_train, X_val, y_val, y_train_pred, y_val_pred)
```

```python
from sklearn.ensemble import GradientBoostingClassifier

gbr = GradientBoostingClassifier(n_estimators=10, learning_rate=0.1).fit(X_train, y_train)
#prediction
y_train_pred=gbr.predict(X_train)
y_val_pred=gbr.predict(X_val)


#visualisation
visualise(X_train, X_val, y_val, y_train_pred, y_val_pred)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import *
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.metrics import RootMeanSquaredError
from tensorflow.keras.optimizers import Adam

simple_nn = Sequential()
simple_nn.add(InputLayer((columns_count,)))
simple_nn.add(Dense(784, 'relu'))
simple_nn.add(Dense(10, 'softmax'))


# print(y_train_onehot.shape, y_val_onehot.shape, y_val_onehot[1:4])
opt = Adam(learning_rate=.01)
cp = ModelCheckpoint('models/models-image/simple_nn.keras', save_best_only=True)
simple_nn.compile(optimizer=opt, loss='mse', metrics=[RootMeanSquaredError()])
simple_nn.fit(x=X_train, y=y_train_onehot, validation_data=(X_val, y_val_onehot), callbacks=[cp], epochs=4)
```

```python
from tensorflow.keras.models import load_model

# simple_nn = load_model('models/models-image/simple_nn.keras')
#prediction
y_train_pred=simple_nn.predict(X_train)
y_val_pred=simple_nn.predict(X_val)

#visualisation
visualise_accuracy(X_train, X_val, y_train_onehot, y_val_onehot, y_train_pred, y_val_pred) 
```
